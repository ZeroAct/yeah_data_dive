# Distribution
You may heard about the word "distribution" at school or work.

In this section, we will dive into what is distribution and how to use it.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's start with a simple example.

We will use 2-d array to represent the distribution for intuitive understanding.

```python
data1 = np.array([-1, -1])
data2 = np.array([1, 1])
```

```python
x = np.array([data1[0], data2[1]])
y = np.array([data1[1], data2[1]])
plt.scatter(x, y, c='r')
```

A set of data can be called a population or sample.

Sample is a subset of population.

The above example is a population with 2 data.

Population has parameters like mean, variance, etc.

We can get the parameters of population by using numpy.

```python
mean = np.mean([data1, data2], axis=0)
var = np.var([data1, data2], axis=0)

print(f"Mean of population: {mean}")
print(f"Variance of population: {var}")
```

Lets add some more data to the population.

```python
datas = np.array([
    [-1, -1],
    [-0.5, -0.5],
    [-1, 1],
    [-0.3, 0.3],
    [1, -1],
    [0.5, -0.5],
    [1, 1],
    [0.3, 0.3]
])
```

```python
x = datas[:, 0]
y = datas[:, 1]
plt.scatter(x, y, c='b')

mean = np.mean(datas, axis=0)
var = np.var(datas, axis=0)

print(f"Mean of sample: {mean}")
print(f"Variance of sample: {var}")
```

So, we can say that

the mean of the population is [2.08166817e-17 -5.00000000e-02],

and the variance of the population is [0.585  0.5825].

This nuance is distribution.

We can sample a data from the distribution.

```python
# Bing copilot generated this code
import math
import random

def generate_random(mu, var):
    """
    Generate a number following a normal distribution with mean mu and variance var
    """
    u1 = random.uniform(0, 1)
    u2 = random.uniform(0, 1)
    z0 = math.sqrt(-2 * math.log(u1)) * math.cos(2 * math.pi * u2)
    return mu + math.sqrt(var) * z0
```

The above function generates a random value that follows given mean and variance using Box-Muller transform algorithm.

```python
x = generate_random(2.08166817e-17, 0.585)
y = generate_random(-5.00000000e-02, 0.5825)

print(f"New data point: {x}, {y}")
```

Lets generate many samples from the distribution we made!

```python
x_generator = lambda N: np.array([generate_random(2.08166817e-17, 0.585) for _ in range(N)])
y_generator = lambda N: np.array([generate_random(-5.00000000e-02, 0.5825) for _ in range(N)])
```

```python
datas = np.array([x_generator(1000), y_generator(1000)]).T

x = datas[:, 0]
y = datas[:, 1]
plt.scatter(x, y, c='b')

mean = np.mean(datas, axis=0)
var = np.var(datas, axis=0)

print(f"Mean of sample: {mean}")
print(f"Variance of sample: {var}")
```

Surprise! We just generated 1000 samples from the distribution we made with only 8 samples.

There are many distributions in the world.

Lets see the most well-known distribution called normal distribution.

Normal distribution has 0 mean and 1 variance.

We can use the above code, but we will use `numpy.random.randn` function for simplicity.

```python
datas = np.random.randn(100, 2)

x = datas[:, 0]
y = datas[:, 1]
plt.scatter(x, y, c='b')

mean = np.mean(datas, axis=0)
var = np.var(datas, axis=0)

print(f"Mean of sample: {mean}")
print(f"Variance of sample: {var}")
```

The above example shows 100 data from normal distribution.

It has [-0.02955945 -0.03790801] mean and [0.73751896 0.95629977] variance.

Lets generate more data.

```python
datas = np.random.randn(1000, 2)

x = datas[:, 0]
y = datas[:, 1]
plt.scatter(x, y, c='b')

mean = np.mean(datas, axis=0)
var = np.var(datas, axis=0)

print(f"Mean of sample: {mean}")
print(f"Variance of sample: {var}")
```

We can see that more data we generate, the more the distribution looks like normal distribution.

This means posterior distribution of normal distribution is normal distribution.

Lets see another example in some real world data.

Assume that there is a table that contains the height, weight and sex of 1000 people.

Male population has 175 cm mean and 20 cm variance in height, 75 kg mean and 30 variacne in weights.

Female population has 160 cm mean and 15 cm variance in height, 55 kg mean and 15 variacne in weights.

```python
male_heights = np.random.normal(175, math.sqrt(20), 100)
male_weights = np.random.normal(75, math.sqrt(30), 100)

female_heights = np.random.normal(160, math.sqrt(15), 100)
female_weights = np.random.normal(55, math.sqrt(15), 100)
```

```python
plt.scatter(male_heights, male_weights, c='b')
plt.scatter(female_heights, female_weights, c='r')

plt.xlabel("height (cm)")
plt.ylabel("weight (kg)")
plt.legend(["male", "female"])
```

As we can see above, there are significant mismatch between male distribution and femail distribution.

Training a model is the process of learning the distribution of data and finding the answer that we intend.

If the distribution of the data changes, the model may not work as expected.

This is why using a model trained on female data to predict the height of male data is not a good choice.

The difference between the distributions causes the model to fail.

This is an example of `covariate shift`, a type of `domain shift` or `dataset shift`.

It occurs when the input distribution changes, but the output distribution remains the same.
